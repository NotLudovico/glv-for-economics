```python
import numpy as np
from glv.visualization import plot_shortest_path_distribution
```

```python
degrees_exp = np.random.exponential(scale=4, size=10000).astype(int)
plot_shortest_path_distribution(degrees_exp, title="Exponential Graph Distribution")
```

```python
n = 10000
degrees_pl = np.random.zipf(a=2.4, size=n)
degrees_pl = np.clip(degrees_pl, 1, n - 1)
plot_shortest_path_distribution(degrees_pl, title="Power-Law Graph Distribution")
```
